```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
pokemon = pd.read_csv('pokemon.csv')
pokemon.head(10)
```

```python
plt.hist(data = pokemon, x = 'speed');
# similar to the bar plot only
# first argument is the data to be read, and second is the x axis argument to be taken into account
# if no argument given matlab by default generate 10 bins
# bin values are not particularly aligned with the tick marks, making interpretation tricky
```

```python
plt.hist(data = pokemon, x = 'speed', bins = 30);
# bins argument tells how many bins are required to show the data
# but still has problem of the bin boundaries as the bins are generated in 20 equal portions
# note in this case bin values are not integer but data values are integers
# so some bin values cover more bin values than others
```

```python
bins = np.arange(0,pokemon['speed'].max()+5, 5)
# to create our own range of bins
# +5 is added as if not added this wouldnt have covered the maximum value
plt.hist(data = pokemon , x = 'speed',  bins=bins);
```

```python
sb.distplot(pokemon['speed']);
# histogram genereated using the seaborn
# only one argument that is the panda series containing all the data
# default bit count is larger than the matplotlib histogram
# a line is also drawn with the y-axis values very small
# line is kernel density estimate (KDE) of data distribution
# total area under the curve is equal to 1
```

```python
sb.distplot(pokemon['speed'], kde = False,  hist_kws = {'alpha' : 1});
# to turn off the density curve use kde = False
```
