```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
two_dice_roll_sum = np.random.randint(1, 7, 100) + np.random.randint(1, 7, 100)
print(two_dice_roll_sum)
```

```python
plt.figure(figsize = [10,3]);

# histogram on the left, bin edges on integers
plt.subplot(1,2,1)
bin_edges = np.arange(2, 12+1.1, 1)
plt.hist(two_dice_roll_sum, bins=bin_edges);
plt.xticks(np.arange(2, 12+1, 1));

# histogram on the left, bin edges on integers
plt.subplot(1,2,2)
bin_edges = np.arange(1.5, 12+0.5, 1)
plt.hist(two_dice_roll_sum, bins=bin_edges);
plt.xticks(np.arange(2, 12+1, 1));
```

```python
plt.figure(figsize = [3.25,3]);
bin_edges = np.arange(1.5, 12.5+1, 1)
plt.hist(two_dice_roll_sum, bins = bin_edges, rwidth = 0.7); # with rwidth = 0/7, the bars will take 70% of the space allocated by each bin, 30% of the space left emplty
plt.xticks(np.arange(2, 12+1, 1));
```
