```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
# mostly seaborn uses the abbreviation as sbs but for this course it is being used as sb

%matplotlib inline 
# helps to render plots directly in the notebook
```

```python
pokemon = pd.read_csv("pokemon.csv")
print(pokemon.shape)
pokemon.head(10)
```

```python
sb.countplot(data = pokemon, x = 'generation_id');
# to generate the plot using the seaborn function sb.countplot(data = dataframe, x = 'parameter of interest')
# ; is used to supress the information of prinitng object
```

```python
sb.color_palette()
# returns the set of tupples of all the colors present in the seaborn color palette
```

```python
base_color = sb.color_palette()[1] 
# to set the base color to a different color stored at the location 1 of the seaborn color pallete
sb.countplot(data = pokemon, x = 'generation_id', color = base_color)
# set color of the bar chart with the base_color
```

```python
sb.countplot(data = pokemon, x = 'generation_id', color = base_color, order = [5,1,3,4,2,7,6])
# a method to sort the bar chart according to the frequency of the each item
```

```python
pokemon['generation_id'].value_counts()
# tells which particular value of generation_id contains how many entries, and sort the pokemon dataframe according to this value in decending order and returns the list 
gen_order = pokemon['generation_id'].value_counts().index
# store the order containing a list of values in ascending order according to the frequency of each item
gen_order
sb.countplot(data = pokemon, x = 'generation_id', order = gen_order)
# plot the bar chart according to the frequency of the each item
```
