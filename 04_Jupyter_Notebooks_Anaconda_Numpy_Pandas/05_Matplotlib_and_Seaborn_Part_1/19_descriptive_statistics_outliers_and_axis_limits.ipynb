```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb


%matplotlib inline
```

```python
pokemon = pd.read_csv('pokemon.csv')
pokemon.head(10)
```

```python
bins = np.arange(0, pokemon['height'].max()+0.5, 0.5)
plt.hist(data = pokemon, x = 'height', bins = bins);
```

```python
bins = np.arange(0, pokemon['height'].max()+0.2, 0.2)
plt.hist(data = pokemon, x = 'height', bins = bins);
plt.xlim((0,6)) # to fix the upper and lower bounds of the x axis
# first argument represents the lower bounds and the second argument represents the upper bound
```
