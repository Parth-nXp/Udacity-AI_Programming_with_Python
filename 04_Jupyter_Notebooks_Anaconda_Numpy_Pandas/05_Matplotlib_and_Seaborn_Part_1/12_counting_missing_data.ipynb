```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
pokemon = pd.read_csv('pokemon.csv')
pokemon.head()
```

```python
na_counts = pokemon.isna().sum()
# pandas functions to create a table with the number of missing values in each column.
base_color = sb.color_palette()[0]
sb.barplot(x=na_counts.index.values, y=na_counts, color=base_color)
# first argument contations the x-values (column names), the second argument the y-values (our counts).
plt.xticks(rotation = 90);
```
