```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
fuel_econ = pd.read_csv("fuel_econ.csv")
print(fuel_econ.shape)
fuel_econ.head(10)
```

```python
plt.scatter(data = fuel_econ , x = "displ", y = "comb");
# we have to explicitly state the x and y data
plt.xlabel("Displacement (l)");
plt.ylabel("Combined Fuel Eff. (mpg)");
```

```python
sb.regplot(data = fuel_econ , x = "displ", y = "comb");
# scatterplot with the fitting of the regression line in the data using seaborn
plt.xlabel("Displacement (l)");
plt.ylabel("Combined Fuel Eff. (mpg)");
```

```python
sb.regplot(data = fuel_econ , x = "displ", y = "comb", fit_reg = False);
# to turn off the regression line set fig_reg = False
plt.xlabel("Displacement (l)");
plt.ylabel("Combined Fuel Eff. (mpg)");
```
