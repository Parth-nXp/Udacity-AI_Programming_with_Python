```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
fuel_econ = pd.read_csv('fuel_econ.csv')
fuel_econ.head()
```

```python
base_color = sb.color_palette()[0];
sb.boxplot(data = fuel_econ, x = 'VClass', y = 'comb', color = base_color);
plt.xticks(rotation = 15);
```

```python
base_color = sb.color_palette()[0];
sb.barplot(data = fuel_econ, x = 'VClass', y = 'comb', color = base_color);
plt.xticks(rotation = 15);
plt.ylabel("Avg. Combined Fuel Eff. (mpg)");
```

```python
base_color = sb.color_palette()[0];
sb.barplot(data = fuel_econ, x = 'VClass', y = 'comb', color = base_color, errwidth=0);
# to set the error wicks to 0 set errwidth = 0
plt.xticks(rotation = 15);
plt.ylabel("Avg. Combined Fuel Eff. (mpg)");
```

```python
base_color = sb.color_palette()[0];
sb.barplot(data = fuel_econ, x = 'VClass', y = 'comb', color = base_color, errorbar = 'sd');
# to change the plot to show standard deviation instead of mean use errorbar = 'sd'
plt.xticks(rotation = 15);
plt.ylabel("Avg. Combined Fuel Eff. (mpg)");
```

```python
sb.pointplot(data = fuel_econ, x = 'VClass', y = 'comb',  errorbar = 'sd');
# to change the plot to point plot instead of the bar plot
plt.xticks(rotation = 15);
plt.ylabel("Avg. Combined Fuel Eff. (mpg)");
```

```python
sb.pointplot(data = fuel_econ, x = 'VClass', y = 'comb',  errorbar = 'sd', linestyles = "");
# to change the plot so that the points are not connected with lines we use linestyles = "" i.e. an empty string
plt.xticks(rotation = 15);
plt.ylabel("Avg. Combined Fuel Eff. (mpg)");
```
