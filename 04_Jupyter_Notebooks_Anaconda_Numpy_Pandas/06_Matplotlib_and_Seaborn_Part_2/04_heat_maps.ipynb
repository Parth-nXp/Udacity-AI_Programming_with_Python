```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
fuel_econ = pd.read_csv('fuel_econ.csv')
fuel_econ.head(10)
```

```python
plt.hist2d(data = fuel_econ, x = 'displ', y = 'comb');
# use matplotlib hist2d to plot the heatmap
plt.xlabel('Displacement (l)');
plt.ylabel('Combined Fuel Eff. (mpg)');
```

```python
plt.hist2d(data = fuel_econ, x = 'displ', y = 'comb');
plt.colorbar();
# to add colorbar to the heat map default one is 'viridis'
plt.xlabel('Displacement (l)');
plt.ylabel('Combined Fuel Eff. (mpg)');

# by default each plot is divided into 10 bins in x and y direction
```

```python
plt.hist2d(data = fuel_econ, x = 'displ', y = 'comb', cmin = 0.5);
# cmin tells about the minimum number of datapoints that needs to be present in any cell to get it colored.
# 0.5 tells that if there are no datapoints in any cell they wont get colored
plt.colorbar();
plt.xlabel('Displacement (l)');
plt.ylabel('Combined Fuel Eff. (mpg)');
```

```python
plt.hist2d(data = fuel_econ, x = 'displ', y = 'comb', cmin = 0.5, cmap = 'viridis_r');
# since ligther color means less datapoint so viridis_r reverses the default colorbar that is viridis
plt.colorbar();
plt.xlabel('Displacement (l)');
plt.ylabel('Combined Fuel Eff. (mpg)');
```

```python
fuel_econ[['displ', 'comb']].describe()
```

```python
bins_x = np.arange(fuel_econ['displ'].min(), fuel_econ['displ'].max()+0.3, 0.3)
bins_y = np.arange(fuel_econ['comb'].min(), fuel_econ['comb'].max()+3, 3)
plt.hist2d(data = fuel_econ, x = 'displ', y = 'comb', cmin = 0.5, cmap = 'viridis_r', bins = [bins_x, bins_y]);
# to set bin sizes, bins take a list with first entry for the x axis and secodn for y axis
plt.colorbar();
plt.xlabel('Displacement (l)');
plt.ylabel('Combined Fuel Eff. (mpg)');
```
