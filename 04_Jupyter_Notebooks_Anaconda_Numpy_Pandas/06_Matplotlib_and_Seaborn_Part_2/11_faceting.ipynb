```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
fuel_econ = pd.read_csv("fuel_econ.csv")
fuel_econ.head()
```

```python
base_color = sb.color_palette()[0];
sb.violinplot(data = fuel_econ, x = 'VClass', y = 'comb', color = base_color, inner = 'quartile');
plt.xticks(rotation = 15);
```

```python
g = sb.FacetGrid(data = fuel_econ, col = 'VClass');
# this means that there will be one plot for each vechile class
g.map(plt.hist, 'comb');
# first parameter to set what plot to use
# second parameter to set what variable to plot
```

```python
bins = np.arange(12, 58+2, 2);
g = sb.FacetGrid(data = fuel_econ, col = 'VClass', col_wrap = 3);
# col_wrap to set how many plots to have in each column
g.map(plt.hist, 'comb', bins = bins);
# third parameter to set the bins for each plot
```

```python
bins = np.arange(12, 58+2, 2);
g = sb.FacetGrid(data = fuel_econ, col = 'VClass', col_wrap = 3, sharey = False);
# sharey = False sets each plot to have its own unique axis
g.map(plt.hist, 'comb', bins = bins);
```
