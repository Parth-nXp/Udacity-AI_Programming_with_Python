```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
fuel_econ = pd.read_csv('fuel_econ.csv')
fuel_econ.head(10)
```

```python
sb.regplot(data = fuel_econ, x = 'year', y = 'comb');
# data discrete on x axis
```

```python
# Jitter
sb.regplot(data = fuel_econ, x = 'year', y = 'comb', x_jitter = 0.3);
# x_jitter means that each x value is randomly adjusted to 0.3 more or 0.3 less
# only works with seaborn regplot function and not with the matplotlib scatter function
```

```python
# Transparency
sb.regplot(data = fuel_econ, x = 'year', y = 'comb', x_jitter = 0.3, scatter_kws= {'alpha': 1/10});

# scatter_kws takes a dictionary as inout with alpha taking value between 0 and 1.
# 0 means fully transparent and 1 means fully opaque
```
